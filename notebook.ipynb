```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 39*np.pi/2, 1000)

x = t * np.cos(t)**3
y = 9*t * np.sqrt(np.abs(np.cos(t))) + t * np.sin(0.2*t ) * np.cos(4*t)
segments = [
    (0, 200, 'green'),# green segments
    (200, 600, 'orange'),# red segments

    (600,1000 ,'red')#orange segments
]

for start, end, color in segments:
    plt.plot(x[start:end], y[start:end], c=color)
plt.show()
   
```
